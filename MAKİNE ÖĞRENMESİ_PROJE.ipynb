```python
from tkinter import *
import tkinter.messagebox as tkMessageBox
import csv, sqlite3
from tkinter.ttk import Combobox

root = Tk()
root.title("Makina Öğrenme Projesi")
 
width = 640
height = 500
screen_width = root.winfo_screenwidth()
screen_height = root.winfo_screenheight()
x = (screen_width/2) - (width/2)
y = (screen_height/2) - (height/2)
root.geometry("%dx%d+%d+%d" % (width, height, x, y))

#VARIABLES
USERNAME = StringVar()
PASSWORD = StringVar()
FIRSTNAME = StringVar()
LASTNAME = StringVar()

AGE = BooleanVar()
JOB = BooleanVar()
MARITAL = BooleanVar()
EDUCATION = BooleanVar()
SORUNLU = BooleanVar()
BALANCE = BooleanVar()
HOUSING = BooleanVar()
LOAN = BooleanVar()
CONTACT = BooleanVar()
DAY = BooleanVar()
MONTH = BooleanVar()

#METHODS
def Database():
    global conn, cursor
    conn = sqlite3.connect("kullanici.db", timeout=10)
    cursor = conn.cursor()
    cursor.execute("CREATE TABLE IF NOT EXISTS kullanicilar (userid INTEGER PRIMARY KEY AUTOINCREMENT NOT NULL, username TEXT, password TEXT, firstname TEXT, lastname TEXT)")

def Exit():
    result = tkMessageBox.askquestion('System', 'Çıkmak istediğinize eminmisiniz?', icon="warning")
    if result == 'yes':
        root.destroy()
        exit()

def LoginForm():
    global LoginFrame, lbl_result1
    LoginFrame = Frame(root)
    LoginFrame.pack(side=TOP, pady=5)
    
    lbl_username = Label(LoginFrame, text="KULLANICI ADI:", font=('arial', 10), bd=5)
    lbl_username.grid(row=1)
    lbl_password = Label(LoginFrame, text="PAROLA:", font=('arial', 10), bd=5)
    lbl_password.grid(row=2)
    lbl_result1 = Label(LoginFrame, text="", font=('arial', 10))
    lbl_result1.grid(row=3, columnspan=2)
    
    username = Entry(LoginFrame, font=('arial', 10), textvariable=USERNAME, width=15)
    username.grid(row=1, column=1)
    
    password = Entry(LoginFrame, font=('arial', 10), textvariable=PASSWORD, width=15, show="*")
    password.grid(row=2, column=1)
    
    btn_login = Button(LoginFrame, text="Login", font=('arial', 10), width=35, command=Login)
    btn_login.grid(row=4, columnspan=2, pady=20)
    
    lbl_register = Label(LoginFrame, text="YENİ KULLANICI EKLE", fg="Blue", font=('arial', 10))
    lbl_register.grid(row=0, sticky=W)
    lbl_register.bind('<Button-1>', ToggleToRegister)
    

def RegisterForm():
    global RegisterFrame, lbl_result2
    
    RegisterFrame = Frame(root)
    RegisterFrame.pack(side=TOP, pady=5)
    
    lbl_username = Label(RegisterFrame, text="KULLANICI ADI:", font=('arial', 10), bd=18)
    lbl_username.grid(row=1)
    lbl_password = Label(RegisterFrame, text="PAROLA:", font=('arial', 10), bd=18)
    lbl_password.grid(row=2)
    lbl_firstname = Label(RegisterFrame, text="ADI:", font=('arial', 10), bd=18)
    lbl_firstname.grid(row=3)
    lbl_lastname = Label(RegisterFrame, text="SOYADI:", font=('arial', 10), bd=18)
    lbl_lastname.grid(row=4)
    lbl_result2 = Label(RegisterFrame, text="", font=('arial', 10))
    lbl_result2.grid(row=5, columnspan=2)
    
    username = Entry(RegisterFrame, font=('arial', 10), textvariable=USERNAME, width=15)
    username.grid(row=1, column=1)
    password = Entry(RegisterFrame, font=('arial', 10), textvariable=PASSWORD, width=15, show="*")
    password.grid(row=2, column=1)
    
    firstname = Entry(RegisterFrame, font=('arial', 10), textvariable=FIRSTNAME, width=15)
    firstname.grid(row=3, column=1)
    lastname = Entry(RegisterFrame, font=('arial', 10), textvariable=LASTNAME, width=15)
    lastname.grid(row=4, column=1)
    
    btn_login = Button(RegisterFrame, text="YENİ KULLANICI EKLE", font=('arial', 10), width=35, command=Register)
    btn_login.grid(row=6, columnspan=2, pady=20)
    
    lbl_login = Label(RegisterFrame, text="LOGIN", fg="Blue", font=('arial', 10))
    lbl_login.grid(row=0, sticky=W)
    lbl_login.bind('<Button-1>', ToggleToLogin)

def ParametreForm():
    global ParametreFrame
    ParametreFrame = Frame(root)
    ParametreFrame.pack(side=TOP)

    label_baslik = Label(ParametreFrame, text="BAĞIMSIZ DEĞİŞKENLER", justify=LEFT)
    label_baslik.grid(row=0)

    Age = Checkbutton(ParametreFrame, text="AGE", var=AGE, justify=LEFT)
    Age.grid(row=1)

    Job = Checkbutton(ParametreFrame, text="JOB", var=JOB, justify=LEFT)
    Job.grid(row=2)

    Marital = Checkbutton(ParametreFrame, text="MARITAL", var=MARITAL, justify=LEFT)
    Marital.grid(row=3)

    Education = Checkbutton(ParametreFrame, text="EDUCATION", var=EDUCATION)
    Education.grid(row=4)

    Education = Checkbutton(ParametreFrame, text="SORUNLU", var=SORUNLU)
    Education.grid(row=5)
    
    Balance = Checkbutton(ParametreFrame, text="BALANCE", var=BALANCE)
    Balance.grid(row=6)
    
    Housing = Checkbutton(ParametreFrame, text="HOUSING", var=HOUSING)
    Housing.grid(row=7)
    
    Loan = Checkbutton(ParametreFrame, text="LOAN", var=LOAN)
    Loan.grid(row=8)
    
    Contact = Checkbutton(ParametreFrame, text="CONTACT", var=CONTACT)
    Contact.grid(row=9)
    
    Day = Checkbutton(ParametreFrame, text="DAY", var=DAY)
    Day.grid(row=10)
    
    Month = Checkbutton(ParametreFrame, text="MONTH", var=MONTH)
    Month.grid(row=11)
    
    btn_hesapla = Button(ParametreFrame, text="HESAPLA", font=('arial', 10), width=35, command=Hesapla)
    btn_hesapla.grid(row=12, pady=20)
    
def ToggleToParametre(event=None):
    LoginFrame.pack_forget()
    ParametreForm()

def ToggleToLogin(event=None):
    RegisterFrame.pack_forget()
    LoginForm()

def ToggleToRegister(event=None):
    LoginFrame.pack_forget()
    RegisterForm()

def Hesapla():
    SECILENLER = ""
    SECILMEYENLER = ""
    
    print(AGE.get(),JOB.get(),MARITAL.get(),EDUCATION.get(), SORUNLU.get(), BALANCE.get(), HOUSING.get(), LOAN.get(), CONTACT.get(), DAY.get(), MONTH.get())
  
    
    lstColumns = []
    
    if AGE.get() == True:
        lstColumns.append('age')
        if SECILENLER == "":
            SECILENLER = "'age'"
        else:
            SECILENLER = SECILENLER + ",'age'"
    else:
        if SECILMEYENLER == "":
            SECILMEYENLER = "'age'"
        else:
            SECILMEYENLER = SECILMEYENLER + ",'age'"
            
    if JOB.get() == True:
        lstColumns.append('job')
        if SECILENLER == "":
            SECILENLER = "'job'"
        else:
            SECILENLER = SECILENLER + ",'job'"
    else:
        if SECILMEYENLER == "":
            SECILMEYENLER = "'job'"
        else:
            SECILMEYENLER = SECILMEYENLER + ",'job'"
            
    if MARITAL.get() == True:
        lstColumns.append('marital')
        if SECILENLER == "":
            SECILENLER = "'marital'"
        else:
            SECILENLER = SECILENLER + ",'marital'"
    else:
        if SECILMEYENLER == "":
            SECILMEYENLER = "'marital'"
        else:
            SECILMEYENLER = SECILMEYENLER + ",'marital'"
            
    if EDUCATION.get() == True:
        lstColumns.append('education')
        if SECILENLER == "":
            SECILENLER = "'education'"
        else:
            SECILENLER = SECILENLER + ",'education'"
    else:
        if SECILMEYENLER == "":
            SECILMEYENLER = "'education'"
        else:
            SECILMEYENLER = SECILMEYENLER + ",'education'"
        
    if SORUNLU.get() == True:
        lstColumns.append('sorunlu')
        if SECILENLER == "":
            SECILENLER = "'sorunlu'"
        else:
            SECILENLER = SECILENLER + ",'sorunlu'"
    else:
        if SECILMEYENLER == "":
            SECILMEYENLER = "'sorunlu'"
        else:
            SECILMEYENLER = SECILMEYENLER + ",'sorunlu'"
        
    if BALANCE.get() == True:
        lstColumns.append('balance')
        if SECILENLER == "":
            SECILENLER = "'balance'"
        else:
            SECILENLER = SECILENLER + ",'balance'"
    else:
        if SECILMEYENLER == "":
            SECILMEYENLER = "'balance'"
        else:
            SECILMEYENLER = SECILMEYENLER + ",'balance'"
        
    if HOUSING.get() == True:
        lstColumns.append('housing')
        if SECILENLER == "":
            SECILENLER = "'housing'"
        else:
            SECILENLER = SECILENLER + ",'housing'"
    else:
        if SECILMEYENLER == "":
            SECILMEYENLER = "'housing'"
        else:
            SECILMEYENLER = SECILMEYENLER + ",'housing'"
        
    if LOAN.get() == True:
        lstColumns.append('loan')
        if SECILENLER == "":
            SECILENLER = "'loan'"
        else:
            SECILENLER = SECILENLER + ",'loan'"
    else:
        if SECILMEYENLER == "":
            SECILMEYENLER = "'loan'"
        else:
            SECILMEYENLER = SECILMEYENLER + ",'loan'"
        
    if CONTACT.get() == True:
        lstColumns.append('contact')
        if SECILENLER == "":
            SECILENLER = "'contact'"
        else:
            SECILENLER = SECILENLER + ",'contact'"
    else:
        if SECILMEYENLER == "":
            SECILMEYENLER = "'contact'"
        else:
            SECILMEYENLER = SECILMEYENLER + ",'contact'"
        
    if DAY.get() == True:
        lstColumns.append('day')
        if SECILENLER == "":
            SECILENLER = "'day'"
        else:
            SECILENLER = SECILENLER + ",'day'"
    else:
        if SECILMEYENLER == "":
            SECILMEYENLER = "'day'"
        else:
            SECILMEYENLER = SECILMEYENLER + ",'day'"
        
    if MONTH.get() == True:
        lstColumns.append('month')
        if SECILENLER == "":
            SECILENLER = "'month'"
        else:
            SECILENLER = SECILENLER + ",'month'"
    else:
        if SECILMEYENLER == "":
            SECILMEYENLER = "'month'"
        else:
            SECILMEYENLER = SECILMEYENLER + ",'month'"
        
    print('SEÇİLENLER:' + SECILENLER)
    print('SEÇİLMEYENLER :' + SECILMEYENLER)
    
    #pandas,numpy ve matplotlib kütüphanelerinin yüklenmesi
    import pandas as pd
    import numpy as np
    get_ipython().run_line_magic('matplotlib', 'inline')

    #data setimizi okutma
    df=pd.read_csv('bank_full.csv', delimiter=';', parse_dates=True)
    
    # Bağımlı ve bağımsız değişkenlerin seçimi
    X = pd.DataFrame(df, columns=lstColumns)
    #X = df.loc[:, n, dtype='object']
    #X = df.loc[:,'age', 'month']
    y = df.loc[:,'y']

    # In[110]:
    X
    # In[111]:

    #Bağımsız ve bağımlı değişkenlerin logistic regresyona göre integer değer almasının sağlanması
    X = pd.get_dummies(X, prefix_sep='_',drop_first=True)
    y = pd.get_dummies(y, prefix_sep='_',drop_first=True)

    # In[112]:

    #y kolonunun 0 veya 1 değerinin alınmasının sağlanması
    y.rename(columns={'yes':'y'},inplace = True)

    # In[113]:
    X

    # In[114]:
    y


    # Eğitim ve Test Setlerine Ayırma
    from sklearn.model_selection import train_test_split
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)


    # Feature Scaling (Normalizasyon değişkeni 0 ile 1 arasına getiriyor)
    from sklearn.preprocessing import StandardScaler
    sc_X = StandardScaler()
    X_train = sc_X.fit_transform(X_train)
    X_test = sc_X.transform(X_test)
    #y_train kategorik bağımlı değişken olduğundan scale etmiyoruz.


    #Logistic Regresion modeli ile verimizi eğitme
    from sklearn.linear_model import LogisticRegression
    clf = LogisticRegression()
    clf.fit(X_train,y_train)


    #Test sonuçlarının tahmini
    y_pred = clf.predict(X_test)


    #Confusion Matrix(Hata Matrisi)
    from sklearn.metrics import confusion_matrix
    cm = confusion_matrix(y_test,y_pred)
    cm

    print("Doğruluk Oranı: %",((cm[0][0]+cm[1][1]) / (cm[0][0]+cm[0][1]+cm[1][0]+cm[1][1]))*100)
        
def HesaplaSQL():
    Database()
    WHERE = ""
    
    if AGE1.get() != "" and AGE2.get() != "":
        if WHERE == "":
            WHERE = " age BETWEEN " + AGE1.get() + " AND " + AGE2.get()
        else:
            WHERE = WHERE + " AND age BETWEEN " + AGE1.get() + " AND " + AGE2.get()
    
    if JOB.get() != "":
        if WHERE == "":
            WHERE = " job ='" + JOB.get() + "'"
        else:
            WHERE =  WHERE + " AND job ='" + JOB.get() + "'"

    if MARITAL.get() != "":
        if WHERE == "":
            WHERE = " marital ='" + MARITAL.get() + "'"
        else:
            WHERE =  WHERE + " AND marital ='" + MARITAL.get() + "'"

    if EDUCATION.get() != "":
        if WHERE == "":
            WHERE = " education ='" + EDUCATION.get() + "'"
        else:
            WHERE =  WHERE + " AND education ='" + EDUCATION.get() + "'"

    if SORUNLU.get() == True:
        if WHERE == "":
            WHERE = " sorunlu ='no'"
        else:
            WHERE =  WHERE + " AND sorunlu ='yes'"
        
    if BALANCE1.get() != "" and BALANCE2.get() != "":
        if WHERE == "":
            WHERE = " balance BETWEEN " + BALANCE1.get() + " AND " + BALANCE2.get()
        else:
            WHERE = WHERE + " AND balance BETWEEN " + BALANCE1.get() + " AND " + BALANCE2.get()

    #kayıtlar filtreliyor
    print("SELECT * FROM must_bilgi WHERE " + WHERE)
    
    if WHERE != "":
        cursor.execute("SELECT * FROM must_bilgi WHERE " + WHERE)
        if cursor.fetchone() is not None:
            data = cursor.fetchall()
            print(data)
        else:
            lbl_result3.config(text="Veri bulunamadı!", fg="orange")

        cursor.close()
        conn.close()
        
def Register():
    Database()
    if USERNAME.get == "" or PASSWORD.get() == "" or FIRSTNAME.get() == "" or LASTNAME.get == "":
        lbl_result2.config(text="LÜTFEN TÜM BİLGİLERİ GİRİNİZ!", fg="orange")
    else:
        cursor.execute("SELECT * FROM kullanicilar WHERE username = ?", (USERNAME.get(),))
        if cursor.fetchone() is not None:
            lbl_result2.config(text="BU KULLANICI ZATEN TANIMLI", fg="red")
        else:
            cursor.execute("INSERT INTO kullanicilar (username, password, firstname, lastname) VALUES(?, ?, ?, ?)", (str(USERNAME.get()), str(PASSWORD.get()), str(FIRSTNAME.get()), str(LASTNAME.get())))
            conn.commit()
            USERNAME.set("")
            PASSWORD.set("")
            FIRSTNAME.set("")
            LASTNAME.set("")
            lbl_result2.config(text="KULLANICI OLUŞTURULDU!", fg="black")
        cursor.close()
        conn.close()
        
def Login():
    Database()
    if USERNAME.get == "" or PASSWORD.get() == "":
        lbl_result1.config(text="LÜTFEN KULLANICI ADI VE PAROLA BİLGİLİRİ DOLDURUNUZ!", fg="orange")
    else:
        cursor.execute("SELECT * FROM kullanicilar WHERE username = ? and password = ?", (USERNAME.get(), PASSWORD.get()))
        if cursor.fetchone() is not None:
            ToggleToParametre()
        else:
            lbl_result1.config(text="KULLANICI ADI VE PAROLA YANLIŞ", fg="red")

LoginForm()

#MENUBAR WIDGETS
menubar = Menu(root)
filemenu = Menu(menubar, tearoff=0)
filemenu.add_command(label="Çıkış", command=Exit)
menubar.add_cascade(label="File", menu=filemenu)
root.config(menu=menubar)


#INITIALIZATION
if __name__ == '__main__':
    root.mainloop()
```
